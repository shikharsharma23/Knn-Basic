## Task 1 - Visualization
Visualize the data stored in `DataSet1.csv`!

```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt

# read data
Z = np.genfromtxt('DataSet1.csv', delimiter=',')
X, y = Z[:,:-1], Z[:,-1]

# plot data
# your code goes here
colors = [i for i in y]
plt.scatter(X[:,0],X[:,1],c=colors)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of data. Coded according to class')
plt.show()
```

### Questions 
 - Why is $k$-nearest neighbor possibly a reasonable approach?

### Answers
 - We can see clearly that data is quite groupable into classes and nearby points share the same class. Hence using knn will make sense

## Task 2 - Training the model
Use k-nearest neighbor classifiers with $k \in \{1,3,5,...,47,49,51\}$. Either use the `kNeigbhors` function from `sklearn` or implement it on your own (bonus points!). In order to estimate the generalization error, use 10-fold cross validation. Visualize your results (plot $k$ vs. error rates).

```python
from sklearn import neighbors
from sklearn.model_selection import cross_val_score

# your code goes here
k_values =[]
cv_errors = []
for i in range(0,26):
    classifier = neighbors.KNeighborsClassifier(n_neighbors=2*i+1)
    scores = cross_val_score(classifier, X, y, cv=10, scoring = 'accuracy')
    cv_errors.append(1 - np.mean(scores))
    k_values.append(2*i+1)    
```

```python
plt.plot(k_values,cv_errors)
plt.xlabel('k')
plt.ylabel('cv error')
plt.title('error rates change as per k for nearest neighbors')
plt.show()
```

### Questions
 - Thinking of model complexity as the ability of the model to fit to noise, what choice of $k$ gives complex models? Why?
 - How does the error develop with the model complexity? Why?

### Answers
 - More value of k more is the number of data points a result will be influenced and hence more complex model and more capability tof it noise after a certain point. Erro intially goes down with increasing k but after certain limit it goes on increasing due to noise from far away points

## Task 4 - $k$-NN in higher dimensions
Add four random features to the data matrix $X$ which are uniformly distributed between 0 an 1. Perform the same steps as before.

```python
feature1 = np.random.uniform(size = len(X))
feature2 = np.random.uniform(size = len(X))
feature3 = np.random.uniform(size = len(X))
feature4 = np.random.uniform(size = len(X))
X = np.concatenate((X,feature1.reshape(200,1)),axis=1)
X = np.concatenate((X,feature2.reshape(200,1)),axis=1)
X = np.concatenate((X,feature3.reshape(200,1)),axis=1)
X = np.concatenate((X,feature4.reshape(200,1)),axis=1)


# your code goes here
k_values =[]
cv_errors = []
for i in range(0,26):
    classifier = neighbors.KNeighborsClassifier(n_neighbors=2*i+1)
    scores = cross_val_score(classifier, X, y, cv=10, scoring = 'accuracy')
    cv_errors.append(1 - np.mean(scores))
    k_values.append(2*i+1)    
    
plt.plot(k_values,cv_errors)
plt.xlabel('k')
plt.ylabel('cv error')
plt.title('error rates change as per k for nearest neighbors')
plt.show()
# your code goes here
```
